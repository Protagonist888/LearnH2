# FIrst working Python loop

* Write my python code that does the following:
* (1) allows user to flag dictionary instance that is not correct 
* (2) allows user to update with a new reference 
* (3) this new reference is flagged as new
* (4) new reference must be verified three times by other users so a count mechanism is initiated
* (5) After being verified three times by other users, the new reference takes the place of the old reference
* (6) The user who provided the reference receives a message that his new reference is verified

```python
# Sample dataset
dataset = [
    {"id": 1, "data": "Lorem ipsum", "reference": "ref123", "is_correct": True, "verification_count": 0},
    {"id": 2, "data": "Dolor sit amet", "reference": "ref456", "is_correct": True, "verification_count": 0},
    {"id": 3, "data": "Consectetur adipiscing", "reference": "ref789", "is_correct": True, "verification_count": 0}
]

# Display the dataset
def display_dataset(dataset):
    for instance in dataset:
        print(f"ID: {instance['id']}")
        print(f"Data: {instance['data']}")
        print(f"Reference: {instance['reference']}")
        print(f"Verified: {instance['is_correct']}")
        print(f"Verification Count: {instance['verification_count']}")
        print("-----")

# User input
print(dataset)
user_id = int(input("Enter the ID of the instance to flag as incorrect: "))
new_reference = input("Enter the new reference: ")

# Flag the instance as incorrect and update the new reference
for instance in dataset:
    if instance["id"] == user_id:
        instance["is_correct"] = False
        instance["reference"] = new_reference
        break

# Verify the new reference
verification_count = 0
for instance in dataset:
    if instance["reference"] == new_reference and not instance["is_correct"]:
        while verification_count < 3:
            verification = input("Verify the new reference (Y/N): ")
            if verification.upper() == "Y":
                instance["verification_count"] += 1
                verification_count += 1
                if instance["verification_count"] == 3:
                    instance["is_correct"] = True
                    instance["verification_count"] = 0
                    print("New reference verified and updated successfully!")
                    # Send a message to the user who provided the reference
                    print("Message: Your new reference has been verified and updated.")
                    break
                else:
                    print(f"Verification {instance['verification_count']} recorded. Waiting for {3 - instance['verification_count']} more verifications...")
            elif verification.upper() == "N":
                break

# Display the updated dataset
print("Updated dataset:")
display_dataset(dataset)
```
